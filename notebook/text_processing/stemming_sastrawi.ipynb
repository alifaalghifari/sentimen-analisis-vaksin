## Stemming with sastrawi

```python
# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
# crate stop word
stop_factory = StopWordRemoverFactory()
data_stop_word = stop_factory.get_stop_words()
stopword = stop_factory.create_stop_word_remover()
```

### Get 50% data for lexicon

```python
length_data = len(data)
length_data
```

```python
# 7007 / 2 = 3502

data_lex = data[: int(length_data/2)]
```

```python
data_lex.tail()
```

```python
stemming_word = []
for w in data_lex['Text']:

    # get stemming of word
    word_stemmer   = stemmer.stem(w)
    
    # delete punctuation of words
    word_clean_punctuation = word_stemmer.translate(str.maketrans('','',string.punctuation))

    # delete stop word of words
    word_stopword = stopword.remove(word_clean_punctuation)

    stemming_word.append(word_stopword)
```

```python
stemming_data = pd.DataFrame(stemming_word, columns=['text_stemming'])
```

## Filtering word in kamus indonesia

```python
# open kamus indonesia
kamus_clean = []

with open('../components/kata-dasar-indonesia.txt') as f:
  kamus = f.read()
  kamus = kamus.split(' ')

  for w in kamus:
  
    rx = re.findall(r'[^\n]+$', w)

    kamus_clean.append(rx[0])
    
def kamus_word(word):
    checked = False
    # kamus = map(clean_word, kamus)
    if word in kamus_clean:
      checked = True
      # AKAR_KATA.append(word)
      return checked, word 
    return checked, word
```

```python
list_clean_words = []
for words in stemming_word:
    list_word = []
    words = words.split(' ')

    for word in words:
        checked, _ = kamus_word(word)
        if checked:
            list_word.append(word)
    
    list_clean_words.append(' '.join(list_word))
```

## Lexicon

```python
lexicon_neg = pd.read_csv('../components/InSet-master/negative.tsv', sep="\t")
lexicon_pos = pd.read_csv('../components/InSet-master/positive.tsv', sep="\t")
```

### Create label each text tweet

```python
# create label with lexicon
label = []
list_lexicon_pos = list(lexicon_pos['word'])
list_lexicon_neg = list(lexicon_neg['word'])

for word in list_clean_words:
  word_split = word.split(' ')
  neg = 0
  pos = 0

  for w in word_split:
    checked, w = kamus_word(w)
    if checked:
      if w in list_lexicon_neg :
      
        index = list_lexicon_neg.index(w)
        
        neg += lexicon_neg['weight'][index]

      if w in list_lexicon_pos :
        index = list_lexicon_pos.index(w)
  
        pos += lexicon_pos['weight'][index]
      
  jumlah = neg + pos

  if(jumlah > 0):
    label.append('positif')
  elif(jumlah < 0):
    label.append('negatif') 
  else :
    label.append('netral') 
```
